```python
import pandas as pd
df = pd.read_csv(r'englishAmharicTigrinya2.csv', encoding ="utf-8")
df1 = df.sample(frac =1)
display(df1)
```

```python
sentence = df1.Tigrinya.values.tolist()
sentencee = df1.English.values.tolist()
sentiment = df1.real_class.values.tolist()
```

```python
sentence_a = np.array(sentence)
sentence_e = np.array(sentencee)
sentiment_a =np.array(sentiment)
```

```python
print(len(sentence_a))
print(len(sentiment_a))
print(sentiment[0:10])
```

```python
Comm = []
for el in sentence_a:
    sent= re.sub(r'[A-Z]+', '', el)
    sent1 = re.sub(r'[a-z]+', '',sent)
    sent2= re.sub(r'[^\w]', ' ', sent1)
    
    Comm.append(sent2)
```

```python
print(len(Comm))
print(Comm[0:10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(Comm, sentiment , test_size = 0.20, random_state = 42, shuffle = True)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.20, random_state = 1, shuffle = True)
print(len(X_train))
print(len(X_val))
print(len(X_test))
```

```python
import time
import torch 
from sentence_transformers import SentenceTransformer
from torch import nn, optim, tensor
from torch.nn import functional as F
encoder = SentenceTransformer('sentence-transformers/LaBSE')
print('Encoding segments...')
```

```python
model = SentenceTransformer('sentence-transformers/LaBSE')
embeddingtr = model.encode(X_train, batch_size = 128, convert_to_numpy = True, show_progress_bar = True)
embeddingts = model.encode(X_test, batch_size = 128, convert_to_numpy = True, show_progress_bar = True)
```

```python
from sklearn.svm import SVC
clf = SVC(gamma ='auto')
clf.fit(embeddingtr, y_train)
y_pred = clf.predict(embeddingts)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf1 = RandomForestClassifier(max_depth=9, random_state=0)
clf1.fit(embeddingtr, y_train)

y_pred1 = clf1.predict(embeddingts)
print(accuracy_score(y_test, y_pred1))
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred1)

print('Confusion Matrix\n')
print(confusion)
```
