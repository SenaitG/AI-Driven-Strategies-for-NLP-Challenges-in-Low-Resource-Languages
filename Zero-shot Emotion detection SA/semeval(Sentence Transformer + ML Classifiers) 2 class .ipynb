```python
import pandas as pd
df = pd.read_csv (r'semval3class/third_set6000.csv')
display(df)
```

```python
dataf= df.sort_values('real_class')
```

```python
import pandas 
pandas.set_option('display.max_rows', None)
display(dataf)
```

```python
negative = dataf[0:863]
neutral = dataf[864:2906]
positive = dataf[2906:]

print(len(negative))
print(len(neutral))
print(len(positive))
```

```python
frames =[negative, positive[0:1000]]
dataset = pd.concat(frames)
```

```python
print(len(dataset))
```

```python
display(dataset)
```

```python
data = dataset.sample(frac=1)
display(data)
```

```python
X = data["sequence"].to_numpy()
y = data["real_class"].to_numpy()

```

```python
print(X)
print(len(X))
```

```python
import numpy as np 
import pandas as pd
from sentence_transformers import SentenceTransformer
import torch 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
model = SentenceTransformer('sentence-transformers/LaBSE')
```

```python
embeddings = model.encode(X, batch_size = 128, convert_to_numpy = True, show_progress_bar = True)
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y)
y_a = label_encoder.transform(y)
print(y_a)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(embeddings, y_a, test_size = 500, train_size = 900, random_state = None, shuffle = True, stratify = None)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from keras.models import Sequential
from keras import layers

input_dim = embeddings.shape[1]
model = Sequential()
model.add(layers.Dense(64, input_dim =input_dim, activation = 'relu'))
model.add(layers.Dense(1, activation ='sigmoid'))
model.compile(loss='binary_crossentropy',
             optimizer ='adam',
             metrics=['accuracy'])
model.summary()
```

```python
hist =model.fit(X_train, y_train, epochs = 100, batch_size = 128, verbose = 1)
```

```python
score,acc  = model.evaluate(X_test, y_test, verbose=1)
print('Test accuracy:', acc)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
log_reg_params = [ {"C":0.01}, {"C":0.1}, {"C":1}, {"C":10} ]
```

```python
log_reg_params = [{"C":0.01}, {"C":0.1}, {"C":1}, {"C":10}]
dec_tree_params = [{"criterion": "gini"}, {"criterion": "entropy"}]
rand_for_params = [{"criterion": "gini"}, {"criterion": "entropy"}]
kneighbors_params = [{"n_neighbors":3}, {"n_neighbors":5}]
naive_bayes_params = [{}]
svc_params = [{"C":0.01}, {"C":0.1}, {"C":1}, {"C":10}]
```

```python
modelclasses = [
    ["log regression", LogisticRegression, log_reg_params],
    ["decision tree", DecisionTreeClassifier, dec_tree_params],
    ["random forest", RandomForestClassifier, rand_for_params],
    ["k neighbors", KNeighborsClassifier, kneighbors_params],
    ["naive bayes", GaussianNB, naive_bayes_params],
    ["support vector machines", SVC, svc_params]
]
```

```python
insights = []
for modelname, Model, params_list in modelclasses:
    for params in params_list:
        model = Model(**params)
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        insights.append((modelname, model, params, score))
```

```python
insights.sort(key=lambda x:x[-1], reverse=True)
for modelname, model, params, score in insights:
    print(modelname, params, score)
```
